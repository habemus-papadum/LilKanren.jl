```julia
using LilKanren
```

```julia
slurp() do q, a, b, c, d, e # <--- Logic Vars
    ( q     ≡ (:♣, :♠) )  ∧
    ( (a,b) ≡ q        )  ∧
    ( c     ≡ (1,d)    )  ∧
    ( d     ≡ (1,c)    )  
end
```

```julia
function appendᵒ(o,front,back)
  ∨(  ## <-- Disjunction

    # The base case 
    ∧(() ≡ front,
       o ≡ back),

    # The recursive case
    fresh(3) do h,t,temp
      ∧((h,t)    ≡ front,
        (h,temp) ≡ o,
        @☾(appendᵒ(temp,t,back)))  # <-- Look Here!
    end)
end
```

```julia
slurp() do q
    appendᵒ(q, 
            ❀(:✌︎, :✒︎, :✂︎, :☼), 
            ❀(:☿,:♆))
end
```

```julia
slurp(maxResults=20) do front, back
    appendᵒ(❀(:✌︎, :✒︎, :✂︎, :☼,:☿,:♆,:✌︎, :✒︎, :✂︎, :☼,:☿,:♆),
            front, 
            back)
end
```
